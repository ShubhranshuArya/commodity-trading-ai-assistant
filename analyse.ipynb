```python
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Import Path from pathlib
from pathlib import Path

# Import yfinance package
import yfinance as yf

# Import pandas package
import pandas as pd
```

```python
# Pull the data for the first security
stock_data_1 = yf.Ticker("NG=F")
stock_data_1.info
```

```python
import datetime as dt
stock_data = stock_data_1.history(period="2y", interval="1d")
```

```python
stock_data
```

```python
stock_data_shift_date = (
            stock_data.index[-1] - pd.DateOffset(days=30)
        )
stock_data_shift_date
```

```python
prediction_data = stock_data[stock_data.index <= stock_data_shift_date]
prediction_data
```

```python
original_data = stock_data[stock_data.index > stock_data_shift_date]
original_data
```
